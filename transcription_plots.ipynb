```python
from IPython.display import Audio

audio_file = "./samples/MAPS_MUS-bk_xmas1_ENSTDkCl.wav"
transcripted_midi = "./results/MAPS_MUS-bk_xmas1_ENSTDkCl.mid"
```

```python
import pretty_midi
import matplotlib.pyplot as plt

pretty_midi.pretty_midi.MAX_TICK = 1e10

p_rolls = []

def plot_piano_roll(midi_data, ax, file_name):
    piano_roll = midi_data.get_piano_roll(fs=4)
    p_rolls.append(piano_roll)
    ax.imshow(piano_roll, aspect='auto', cmap='inferno')
    ax.set_xlabel('Time')
    ax.set_ylabel('Note')
    ax.set_xticks([0, piano_roll.shape[1]])
    ax.set_yticks([21,108])
    ax.set_title(file_name)

file_path1 = './samples/MAPS_MUS-bk_xmas1_ENSTDkCl.mid'
file_path2 = transcripted_midi

fig, axs = plt.subplots(2,1, figsize=(12, 6))

# Plot first MIDI file
midi_data1 = pretty_midi.PrettyMIDI(file_path1)
plot_piano_roll(midi_data1, axs[0],'Piano Roll of MAPS_MUS-bk_xmas1_ENSTDkCl Original')

# Plot second MIDI file
midi_data2 = pretty_midi.PrettyMIDI(file_path2)
plot_piano_roll(midi_data2, axs[1],'Piano Roll of MAPS_MUS-bk_xmas1_ENSTDkCl Transcripted')

plt.tight_layout()
```

```python
from music21 import converter

midi_stream = converter.parse(transcripted_midi)

# Iterate through notes and chords in the MIDI file
for element in midi_stream.recurse():
    if 'Note' in element.classes:       # Check if it's a Note
        print(f"Note: {element.nameWithOctave} - Start Time: {element.offset} - Duration: {element.duration.quarterLength}")
    elif 'Chord' in element.classes:    # Check if it's a Chord
        notes_in_chord = ' '.join(n.nameWithOctave for n in element.pitches)
        print(f"Chord: {notes_in_chord} - Start Time: {element.offset} - Duration: {element.duration.quarterLength}")
```
